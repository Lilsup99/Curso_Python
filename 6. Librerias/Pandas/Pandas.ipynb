Pandas es una libreria de Python que su principal caracteriztica es que nos
proporciona un *objeto* llamadado DataFrame, el cual es una estructura de datos.
Pandas permite manipular tablas de datos tabulares (por ejem. una hoja de
excel), nos ayuda a leer y analizar grandes volumenes de dato, aplicar
operaciones matematicas con estos datos, algunos de los objetivos a la
hora de trabajar con Pandas es la extraccion de datos, transformacion de datos,
analisis de datos para obtener conocimientos a partir de estos.

```python
# el siguiente paso es importar la libreria

import pandas as pd
```

Entonces ya sabemos que lo principal que nos ofrece Pandas son los
objetos dataframe. Veamos como crear uno, primero se creara un dataframe a partir de un diccionario

```python
passengers = {"name":[
            "Braund, Mr. Owen Harris",
            "Allen, Mr. William Henry",
            "Bonnell, Miss. Elizabeth"],

             "age": [22, 35, 58],

            "Sex": ["male", "male", "female"]}

passengers
```

```python
#veamos como se ve nuestro diccionario de pasajeros si lo pasamos a un df

df = pd.DataFrame(passengers)

print(type(df))

df
```

Para crear un dataframe de forma manual, lo usual es hacerlo con un diccionario
donde las keys seran los nombres de las columnas del df, y los values seran los datos del df.

### Pandas series

ahora que ya creamos dataframe, veamos el concepto de que es una serie de pandas

```python
# Ahora que tenemos un df, veamos como seleccionar una columna
df['age']
```

```python
#cada columna de un dataframe de pandas se le denomina una pandas series

type(df['name'])
```

![](_rsc\pandas_sereis.png)

```python
# ya vimos como acceder a una columna, ahora veamos como acceder a una fila
# para eso tenemos un metodo llamado loc()

df.loc[1] # aqui accedemos a la fila correspondiente al indice 1
```

```python
# tambien con loc() podemos acceder a un registro en especifico
# primero le pasamos el indice de la fila y luego el indice de la columna

df.loc[1][1]
```
