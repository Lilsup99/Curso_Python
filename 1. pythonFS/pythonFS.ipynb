### Pero que es una variable?

variable en programación es un contenedor que se utiliza para almacenar y manipular datos. Es un nombre simbólico que representa un valor en la memoria de la computadora. Las variables pueden contener diferentes tipos de datos, como números, texto, booleanos, listas u objetos más complejos.
Las variables son fundamentales en la programación, ya que permiten almacenar información que se utilizará en el desarrollo de un programa. Al asignar un valor a una variable, se reserva un espacio en la memoria para almacenar ese valor.

### Ejemplo de condicional if

aqui empezamos a subir un poco el nivel de complejidad, haciendo uso de una sentencia condicional, mas adelante profundizaremos mas en el estudio de este tipo de estructuras.  El siguiente codigo hace uso del condicial if para comprobar si una variable z es igual

# complex
En pocas palabras, los números complejos son aquellos que tienen dos partes:

* Una parte real, como por ejemplo 3 o 7.5.  
* Y otra imaginaria, como 5j o -7j.

```python
a = 5 + 5j
b = 1.3 - 7j

'''
Los números complejos son muy utilizados en las telecomunicaciones y electrónica, 
ya que son muy útiles para describir las ondas electromagnéticas y la corriente eléctrica.
También son usados en diferentes dominios de las matemáticas y en física, 
donde destaca su uso en la mecánica cuántica.
'''

print(a)
print(type(b))
```

# Cadenas (strings)
Las cadenas en Python o strings son un tipo  que permite almacenar secuencias de caracteres. Para crear una, es necesario incluir el texto entre comillas dobles " " o tambien es valido con comillas simples ' '

```python
s = "Esto es una cadena"
print(s)       
print(type(s)) 
```

```python
s = 'Esto es otra cadena'
print(s)        
print(type(s))  
```

Las cadenas no están limitadas en tamaño, por lo que el único límite es la memoria de tu ordenador. Una cadena puede estar también vacía.

```python
s = ''

print(type(s))
```

Tal vez queramos declarar una cadena que contenga variables en su interior, como números o incluso otras cadenas. Una forma de hacerlo sería concatenando la cadena que queremos con otra usando el operador +. Nótese que str() convierte en string lo que se pasa como parámetro.

```python
x = 5
s = "mundo " + str(x)

print(s)
```

Una forma un poco más moderna de realizar lo mismo, es haciendo uso de format( ).

```python
s = "Los números son {} y {}".format(5, 10)
print(s)
```

Es posible también darle nombre a cada elemento, y format( ) se encargará de reemplazar todo.

```python
s = "Los números son {a} y {b}".format(a=5, b=10)
print(s) 
```

## Recorrer una cadena
Las cadenas son cuantitativamente diferentes de los otros dos porque estan hechas de piezas
menores: caracteres. los datos con estas caracteristicas se llaman tipos de datos compuestos.

```python
a = 'metallica'

# podemos tratar a nuestra cadena como una unica cosa
print(a)

# Podemos acceder a los caracteres de una cadena usando notacion de indices (a[i])
print(a[0]) # primera letra, en python siempre se cuenta desde cero
print(a[8]) # Ultima letra.
```

```python
# recorrer todos los caracteres de una cadena

for caracter in a:
    print(caracter)
```

```python
'''
Una juguetería tiene mucho éxito en dos de sus productos: payasos y muñecas. Suele hacer venta por correo y la empresa de logística 
les cobra por peso de cada paquete así que deben calcular el peso de los payasos y muñecas que saldrán en cada paquete a demanda. 
Cada payaso pesa 112 g y cada muñeca 75 g. Escribir un programa que lea el número de payasos y muñecas vendidos en el último pedido 
y calcule el peso total del paquete que será enviado.
'''

peso_pa = 112
peso_mu = 75

n = int(input('cantidad de payasos: '))
m = int(input('cantidad de mu;ecas : '))


peso_total = n * peso_pa + m * peso_mu

print('el peso total es: ', peso_total)
```
